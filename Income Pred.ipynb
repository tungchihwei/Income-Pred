## 2.2. Fill in Missing Values

```python
import pandas as pd
from IPython.display import Markdown, display
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
import numpy as np
from sklearn import metrics

from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import cross_val_score

from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
```

```python
# creat one hot encoder

def oneHotEncoder(data):
    data_col = data.select_dtypes('category').columns # cols that need one hot
    data_1 = data.drop(columns = data_col)
    data_2 = pd.get_dummies(data[data_col])
    
    data = pd.concat([data_1, data_2], axis = 1)
    
    return data
```

### 2.2.1. Workclass

```python
train_data = adult[(adult.workclass.values != ' ?')].copy()
train_label = train_data.workclass

test_data = adult[(adult.workclass.values == ' ?')].copy()
test_label = test_data.workclass

test_data.drop(columns = ['workclass'], inplace = True)
train_data.drop(columns = ['workclass'], inplace = True)

train_data = oneHotEncoder(train_data)
test_data = oneHotEncoder(test_data)

train_data.head()
```

```python
test_data.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

d_tree = DecisionTreeClassifier().fit(train_data, train_label)
d_tree_pred = d_tree.predict(test_data)

ran = RandomForestClassifier().fit(train_data, train_label)
ran_pred = ran.predict(test_data)

log_reg = LogisticRegression().fit(train_data, train_label)
log_reg_pred = log_reg.predict(test_data)
```

```python
major_class = adult.workclass.value_counts().index[0]

pred_workclass = []
for i in range(len(d_tree_pred)):
    com = pd.Series([d_tree_pred[i], ran_pred[i], log_reg_pred[i]])
    
    if com.value_counts()[0] > 1:
        pred_workclass.append(com.value_counts().index[0])
    else:
        pred_workclass.append(major_class)

pred_workclass = pd.Series(pred_workclass)

adult.loc[adult.workclass.values == ' ?', 'workclass'] = pred_workclass.values

print(adult.workclass.value_counts())
```

```python
adult['workclass'] = adult['workclass'].cat.remove_categories(' ?')
```

# 4. Prediction

```python
from sklearn.cross_validation import train_test_split

adult_data = adult.drop(columns = ['income'])
adult_label = adult.income

adult_data = oneHotEncoder(adult_data)
```

```python
adult_train, adult_test, adult_train_label, adult_test_label = train_test_split(adult_data, adult_label, test_size = 0.3)

print("training data number: ", len(adult_train))
print("testing data number: ", len(adult_test))
```

## 4.1. Decision Tree

```python
# Gini 
dtree_gini = tree.DecisionTreeClassifier(criterion = 'gini', max_depth = 10)
dtree_gini = dtree_gini.fit(adult_train, adult_train_label)
dtree_gini_pred = dtree_gini.predict(adult_test)
dtree_gini_acc = accuracy_score(adult_test_label, dtree_gini_pred)
print("Gini Accuracy: %.4f%%" % (dtree_gini_acc*100))

# Entropy
dtree_entropy = tree.DecisionTreeClassifier(criterion = 'entropy', max_depth = 10)
dtree_entropy = dtree_entropy.fit(adult_train, adult_train_label)
dtree_entropy_pred = dtree_entropy.predict(adult_test)
dtree_entropy_acc = accuracy_score(adult_test_label, dtree_entropy_pred)
print("Entropy Accuracy: %.4f%%" % (dtree_entropy_acc*100))

```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt

cm = confusion_matrix(adult_test_label, dtree_gini_pred)
cm
```

## 4.2. Random Forest

```python
# Gini
r_forest_gini = RandomForestClassifier(n_estimators=100, criterion = 'gini', max_features = None)
r_forest_gini.fit(adult_train, adult_train_label)
r_forest_gini_pred = r_forest_gini.predict(adult_test)
r_forest_gini_acc = accuracy_score(adult_test_label, r_forest_gini_pred)
print("Gini Accuracy: %.4f%%" % (r_forest_gini_acc*100))


# Entropy
r_forest_entropy = RandomForestClassifier(n_estimators=100, criterion = 'entropy', max_features = None)
r_forest_entropy.fit(adult_train, adult_train_label)
r_forest_entropy_pred = r_forest_entropy.predict(adult_test)
r_forest_entropy_acc = accuracy_score(adult_test_label, r_forest_entropy_pred)
print("Entropy Accuracy: %.4f%%" % (r_forest_entropy_acc*100))
```

## 4.3. K-Nearest Neighbor 

```python
from sklearn.neighbors import KNeighborsClassifier
max_acc = 0 
for i in range(1, 50):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(adult_train, adult_train_label)
    knn_pred = knn.predict(adult_test)
    knn_acc = accuracy_score(adult_test_label, knn_pred)
    #max_acc = max(max_acc, knn_acc)
    #print(knn_pred)
    print("n=", i, " Accuracy: %.4f%%" % (knn_acc*100))
```

## 4.4. Support Vector Machines

```python
# rbf kernal
svm_clf_rbf = svm.SVC(kernel = 'rbf')
svm_clf_rbf.fit(adult_train, adult_train_label)
svm_clf_rbf_pred = svm_clf_rbf.predict(adult_test)
svm_rbf_score = accuracy_score(adult_test_label, svm_clf_rbf_pred)
print("rbf Accuracy: %.4f%%" % (svm_rbf_score*100))

# Linear kernel
svm_clf_linear = svm.SVC(kernel = 'linear')
svm_clf_linear.fit(adult_train, adult_train_label)
svm_clf_linear_pred = svm_clf_linear.predict(adult_test)
svm_clf_linear_score = accuracy_score(adult_test_label, svm_clf_linear_pred)
print("linear Accuracy: %.4f%%" % (svm_clf_linear_score*100))

# Poly kernal
svm_clf_poly = svm.SVC(kernel = 'poly')
svm_clf_poly.fit(adult_train, adult_train_label)
svm_clf_poly_pred = svm_clf_poly.predict(adult_test)
svm_clf_poly_score = accuracy_score(adult_test_label, svm_clf_poly_pred)
print("poly Accuracy: %.4f%%" % (svm_clf_poly_score*100))

svm_clf_sigmoid = svm.SVC(kernel = 'sigmoid')
svm_clf_sigmoid.fit(adult_train, adult_train_label)
svm_clf_sigmoid_pred = svm_clf_sigmoid.predict(adult_test)
svm_clf_sigmoid_score = accuracy_score(adult_test_label, svm_clf_sigmoid_pred)
print("sigmoid Accuracy: %.4f%%" % (svm_clf_sigmoid_score*100))
```
